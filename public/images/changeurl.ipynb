```python
import pandas as pd
```

```python
data = pd.read_csv("all_recipe.csv", encoding='utf-8')
```

```python
import os
import urllib.request
from urllib.parse import urlparse

for i in range(len(data)):
    url = data['이미지경로'][i]
    parsed_url = urlparse(url)
    filename, file_extension = os.path.splitext(parsed_url.path)

    # 이미지 요청 및 다운로드
    if file_extension.lower() not in ['.jpg', '.jpeg', '.png']:
        file_extension = '.png'
    
    urllib.request.urlretrieve(url, f"{data['메뉴명'][i]}{file_extension}")
    data['이미지경로'][i] = f"{data['메뉴명'][i]}{file_extension}"
```
